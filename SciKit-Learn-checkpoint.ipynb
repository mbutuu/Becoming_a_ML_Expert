## Intro to scikit-learn

```python
#1. Get the data ready
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#create x(features matrix)
x = heart_disease.drop("target", axis=1)
#create y (labels)
y = heart_disease["target"]
```

```python
#2.choose the right model and hyper parameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
#we'll keep the default hyper parameters
clf.get_params()
```

```python
#3.fit the model to the training data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
clf.fit(x_train,y_train);
```

```python
#make a prediction
y_label = clf.predict(np.array([0,2,3,4]))
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
#evaluate the model
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds) 
```

```python
accuracy_score(y_test,y_preds)
```

```python
#5. improve model
#try different amount of n_estimators
np.random.seed(42)
for i in range (10,100,10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=1).fit(x_train,y_train)
    print(f"Model accuracy on test set:{clf.score(x_test,y_test) * 100:.2f}%")
    print("")
```

```python
#6. save model and load it
import pickle
pickle.dump(clf,open("random_forest.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest.pkl","rb"))
loaded_model.score(x_test,y_test)
```

```python
##option 1
missing_car_sales["Make"] = missing_car_sales["Make"].fillna("missing")
missing_car_sales["Odometer (KM)"] = missing_car_sales["Odometer (KM)"].fillna(missing_car_sales["Odometer (KM)"].mean())
missing_car_sales["Colour"] = missing_car_sales["Colour"].fillna("missing")
missing_car_sales["Doors"] = missing_car_sales["Doors"].fillna(4)
```
