## Make a plot from a dictionary

```python
games = {"San-Andreas": 10,
        "GTA 5":9.5,
        "Vice-City":7,
        "Liberty-City":8}
fig, ax = plt.subplots()
ax.bar(games.keys(),games.values());
ax.set(title="Favorite Games",
      xlabel="Games",
      ylabel="Rating");
```

```python
fig, ax = plt.subplots()
ax.barh(list(games.keys()),list(games.values()));
ax.set(title="Favorite Games",
      xlabel="Rating",
      ylabel="Games");
```

```python
carsales= pd.read_csv("car-sales.csv")
carsales
```

```python
carsales["Price"] = carsales["Price"].replace({"//$":"", ",":""}, regex=True).astype(float)
carsales
```

```python
carsales.plot.hist(subplots= True, bins=20);
```

```python
fig, ax = plt.subplots(figsize=(10,6))
carsales.plot(kind="scatter",
         x="Price",
         y="Odometer (KM)",
         c="Doors",
         ax=ax)
```

```python
ax.set_xlim([5500,22500])
```
